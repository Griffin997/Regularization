```python
#%matplotlib widget
import scipy
from scipy.special import iv
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from matplotlib import colors
import numpy as np
from tqdm import trange
from datetime import date
import functools
```

```python
dTE = 8
n_elements = 128
TDATA = np.arange(0, (n_elements+1)*dTE, dTE)

c1=0.3
c2=0.7
T21=45
T22=100
true_params = [c1, c2, T21, T22]

SNR=150
sigma=(c1+c2)/SNR
noise_floor=np.sqrt(np.pi/2)*sigma

iterations = 1000

ParamLabel = [r'$c_1$', r'$c_2$', r'$T_{21}$', r'$T_{22}$']
```

```python
def G_biX_off(t, con_1, con_2, tau_1, tau_2, offSet): 
    signal = con_1*np.exp(-t/tau_1) + con_2*np.exp(-t/tau_2) + offSet
    return signal

def G_biX(t, con_1, con_2, tau_1, tau_2): 
    function = con_1*np.exp(-t/tau_1) + con_2*np.exp(-t/tau_2)
    return function

def expRice(tvect,c1,c2,T21,T22):
    alpha=(G_biX(tvect, c1, c2, T21, T22)/(2*sigma))**2
    Expectation = sigma*np.sqrt(np.pi/2)*np.exp(-alpha)*((1+2*alpha)*scipy.special.ive(0, alpha) + 2*alpha*scipy.special.ive(1,alpha))
    return Expectation

def obj_Rice(params, tvect, noisy_data):
    c1, c2, T21, T22 = params
    riceExpectation = expRice(tvect,c1,c2,T21,T22)
    denoised_data = noisy_data - riceExpectation
    riceFit = G_biX(tvect, c1, c2, T21, T22)
    L2_error_norm = np.linalg.norm(riceFit - denoised_data, 2)**2
    return L2_error_norm

def recreate_sig(tvect, params):
    c1, c2, T21, T22 = params
    riceExpectation = expRice(tvect,c1,c2,T21,T22)
    recreated_sig = G_biX(tvect, c1, c2, T21, T22)
    return recreated_sig + riceExpectation

def obj_McGibney(params, tvect, noisy_data):
    c1, c2, T21, T22 = params
    paramFit = G_biX(tvect, c1, c2, T21, T22)
    noise_offset = 2*sigma**2
    L2_error_norm = np.linalg.norm(noisy_data**2 - paramFit**2 - noise_offset, 2)**2
    return L2_error_norm

def obj_Gudbjartsson(params, tvect, noisy_data):
    c1, c2, T21, T22 = params
    corrected_data = np.abs(noisy_data**2 - sigma**2)**(1/2)
    paramFit = G_biX(tvect, c1, c2, T21, T22)
    L2_error_norm = np.linalg.norm(corrected_data - paramFit, 2)**2
    return L2_error_norm

def add_rician_noise(data, sigma):
    noise_real=np.random.normal(loc=0.0, scale=sigma, size=len(TDATA))
    noise_imag=np.random.normal(loc=0.0, scale=sigma, size=len(TDATA))

    noisy_sig_channel = data + noise_real
    noisy_quadrature_channel = noise_imag

    riceData = np.sqrt(noisy_sig_channel**2 + noisy_quadrature_channel**2)
    return riceData

def check_param_order(popt):
    #Reshaping of array to ensure that the parameter pairs all end up in the appropriate place - ensures that T22 > T21
    if (popt[-1] < popt[-2]): #We want by convention to make sure that T21 is <= T22
        for pi in range(np.size(popt)//2):
            p_hold = popt[2*pi]
            popt[2*pi] = popt[2*pi+1]
            popt[2*pi+1] = p_hold
    return popt
```

```python
data = G_biX(TDATA,*true_params)
```

```python
rician_params = []
mcG_params = []
gud_params = []
offset_params = []
standard_params = []

rician_RSS = []
mcG_RSS = []
gud_RSS = []
offset_RSS = []
standard_RSS = []

for i in trange(iterations):    
    rician_data = add_rician_noise(data, sigma)

    G_biX_results_rice = scipy.optimize.minimize(functools.partial(obj_Rice, tvect = TDATA, noisy_data = rician_data), [0.4,0.6,40,80], method = 'Nelder-Mead')
    rician_params.append(check_param_order(G_biX_results_rice.x))
    rician_RSS.append(G_biX_results_rice.fun)

    G_biX_results_gud = scipy.optimize.minimize(functools.partial(obj_Gudbjartsson, tvect = TDATA, noisy_data = rician_data), [0.4,0.6,40,80], method = 'Nelder-Mead')
    gud_params.append(check_param_order(G_biX_results_gud.x))
    gud_RSS.append(G_biX_results_gud.fun)

    G_biX_results_mcG = scipy.optimize.minimize(functools.partial(obj_McGibney, tvect = TDATA, noisy_data = rician_data), [0.4,0.6,40,80], method = 'Nelder-Mead')
    mcG_params.append(check_param_order(G_biX_results_mcG.x))
    mcG_RSS.append(G_biX_results_mcG.fun)

    G_biX_off_params, _, off_info, _, _ = curve_fit(G_biX_off, TDATA, rician_data, p0=[0.4,0.6,40,80,0.1], full_output = True)
    params_hold = check_param_order(G_biX_off_params[0:-1])
    offset_params.append([*params_hold, G_biX_off_params[-1]])
    offset_RSS.append(np.sum(off_info['fvec']**2))

    G_biX_params, _, info, _, _ = curve_fit(G_biX, TDATA, rician_data, p0=[0.4,0.6,40,80], full_output = True)
    standard_params.append(check_param_order(G_biX_params))
    standard_RSS.append(np.sum(info['fvec']**2))
```

```python
rician_params = np.array(rician_params)
offset_params = np.array(offset_params)
standard_params = np.array(standard_params)
gud_params = np.array(gud_params)
mcG_params = np.array(mcG_params)
```

```python
fig = plt.figure()
plt.plot(TDATA, rician_data, "o", markersize=3, label = 'Data with Rician noise')
plt.plot(TDATA, recreate_sig(TDATA, np.mean(rician_params, axis=0)), color = 'orange', label = 'Rician Expectation with Avg Rician Params')
# plt.plot(TDATA, G_biX(TDATA, *np.mean(rician_params, axis=0)), color = 'green', label = 'Underlying Model with Avg Rician Params')
plt.axhline(y = noise_floor, color = 'r', linestyle = '-', label = 'Noise floor')
plt.xlabel("time")
plt.xlabel("Measurement times", fontsize=15)
plt.ylabel('Amplitudes',fontsize=15)
plt.legend()
```
